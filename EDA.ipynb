#### Step 2: Checking missing values and handling them

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
#Handling it with median values as they are all numerical values

df['LB'].fillna(df['LB'].median(),inplace=True)
import warnings
warnings.filterwarnings('ignore')
```

```python
df['AC'].fillna(df['AC'].median(),inplace=True)
```

```python
df['DS'].fillna(df['DS'].median(),inplace=True)
```

```python
df['DP'].fillna(df['DP'].median(),inplace=True)
```

```python
df['MLTV'].fillna(df['MLTV'].median(),inplace=True)
```

```python
df['Width'].fillna(df['Width'].median(),inplace=True)
```

```python
df['Tendency'].fillna(df['Tendency'].median(),inplace=True)
```

```python
df['NSP'].fillna(df['NSP'].median(),inplace=True)
```

```python
df.isnull().sum()
```

#### Step 3: Detecting outliers - but outliers should not be treated as this is a medical record and outliers could mean emergency

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

outliers = df < (Q1 - 1.5 * IQR) | (df > (Q3 + 1.5 * IQR))
outliers.sum()
```

#### Step 4: Analysis (Data Visualization)

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(16,12), bins=20)
plt.tight_layout()
plt.show()
```

```python
df.boxplot(figsize=(18,6))
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(df['MSTV'], df['MLTV'])
plt.xlabel('MSTV')
plt.ylabel('MLTV')
plt.title('MSTV vs MLTV')
plt.show()
```

```python
plt.scatter(df['ASTV'], df['NSP'])
plt.xlabel('ASTV')
plt.ylabel('NSP')
plt.title('ASTV vs Fetal State (NSP)')
plt.show()
```

#### Step 5: Correlations and heatmaps

```python
import seaborn as sns

plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of CTG Features')
plt.show()
```

```python
df.corr()
```

```python
selected_cols = ['LB','ASTV','MSTV','MLTV','Width','Tendency','NSP']

sns.pairplot(df[selected_cols], hue='NSP', palette='coolwarm', diag_kind='kde')
plt.show()
```
